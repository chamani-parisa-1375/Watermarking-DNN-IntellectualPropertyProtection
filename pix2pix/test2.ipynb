```python
import os 
import numpy as np 
import shutil

s_path = fr'datasets\new_noise_dataset'
d_path = fr'datasets\sub_noise_dataset'



for m,n in [('val',30),('test',30),('train',240)]:

    sm_path = rf'{s_path}/{m}'
    dm_path = rf'{d_path}/{m}'
    os.makedirs(dm_path,exist_ok=True)

    files = os.listdir(sm_path)
    N = len(files)

    rand_chose = np.random.randint(0,N,n)

    for i in rand_chose:

        shutil.copy(rf'{sm_path}/{files[i]}',rf'{dm_path}/{files[i]}')


    
    
```
